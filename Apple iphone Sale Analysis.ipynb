
#sales in india


```python
# Change the column name to 'Product Name'
highest_rated = dp.sort_values(by='Star Rating', ascending=False)
highest_rated = highest_rated.head(10)
print(highest_rated['Product Name'])
```

```python
iphone=highest_rated['Product Name'].value_counts()
iphone.plot(kind='bar')
```

```python
iphone=highest_rated['Product Name'].value_counts()
labels=iphone.index
counts=highest_rated['Number Of Ratings']
fig=px.bar(highest_rated,x=labels,y=counts,title="Number of rating of Highest rated Iphone")
fig.show()
```

```python
iphone=highest_rated['Product Name'].value_counts()
labels=iphone.index
counts=highest_rated['Number Of Reviews']
fig=px.bar(highest_rated,x=labels,y=counts,title="Number of Reviews of Highest rated Iphone")
fig.show()
```

```python
fig=px.scatter( data_frame=dp , x='Number Of Ratings',y='Sale Price',size='Discount Percentage',
               trendline='ols',title='Relationship between discount percentage and Number of Rating ')
fig.show()
```

```python
fig=px.scatter(data_frame=dp, x='Number Of Ratings' , y= 'Discount Percentage',size='Sale Price',
               trendline='ols',title='Relationship between discount percentage and Number Of Rating')
fig.show()
```

```python
labels=highest_rated['Product Name']
count=highest_rated['Number Of Ratings']
fig=px.pie(highest_rated,values=count,names=labels,title='Pie Graph')
fig.show()
```

```python
labels=highest_rated['Product Name']
count=highest_rated['Number Of Ratings']
fig=px.pie(highest_rated,values=count,names=labels,title='Pie Graph')
#explode=[0, 0.3 , 0.2 ,0.6 ,0.2,0.4,0.1,0.2,0.3,0.5]
fig.update_traces(textposition='inside',textinfo='percent+label',marker=dict(line=dict(color='#000000', width=2)))
fig.show()
```

```python
labels=highest_rated['Product Name']
count=highest_rated['Number Of Ratings']
fig=px.pie(highest_rated,values=count,names=labels,title='Pie Graph')
#explode=[0, 0.3 , 0.2 ,0.6 ,0.2,0.4,0.1,0.2,0.3,0.5]
fig.update_traces(textposition='inside',textinfo='percent+label',marker=dict(line=dict(color='#000000', width=2)))
fig.update_traces(hole=.4)
fig.show()
```
